# loc 속성으로 행 단위 데이터 추출하기

```python
import pandas as pd
df=pd.read_csv("../data/gapminder.tsv", sep="\t")

print(df.loc[0])
print("")
print(df.loc[90])
print("")

number_of_rows=df.shape[0]
last_row_index=number_of_rows-1
print(df.loc[last_row_index])
print("")

print(df.loc[[1,90]],df.tail(n=1))
```

# iloc 속성으로 행 단위 데이터 추출하기

```python
print(df.iloc[0])
print("")
print(df.iloc[90])
print("")
print(df.iloc[-1])
print("")
print(df.iloc[[0,90,-1]])

#df.loc[행,렬]이나 df.iloc[행,렬] 도 가능하다.
```

# 파이썬 슬라이싱 구문을 조합하여 원하는 데이터 추출하기

```python
subset=df.loc[:,["year","pop"]]
print(subset.head())
print("")
print(subset.tail())
print("")

subset_iloc=df.iloc[:,[2,4,-1]]
print(subset_iloc.head())
print("")
print(subset_iloc.tail())
```

# iloc 속성과 range 메서드로 원하는 데이터 추출하기

```python
small_range=list(range(5))
print(small_range)
print(type(small_range))
print("")

subset=df.iloc[:,small_range]
print(subset.head())
print("")

small_range=list(range(3,6))
print(small_range)
subset=df.iloc[:,small_range]
print(subset.head())
print("")

small_range=list(range(0,6,2))
print(small_range)
subset=df.iloc[:,small_range]
print(subset.head())
print(subset.tail())
print("")
```

# 열 지정값에 파이썬 슬라이싱을 사용하여 원하는 데이터 추출하기

```python
import pandas as pd
df=pd.read_csv("../data/gapminder.tsv", sep="\t")

subset=df.iloc[:,:3]
print(subset)
print("")
print(subset.head())
print(subset.tail())
print("")

subset=df.iloc[:,0:6:2]
print(subset.head())
print(subset.tail())
```

# loc, iloc 자유자재로 사용하기

```python
#행, 렬을 입력하여 자유자재로 출력
print(df.iloc[[0,90,900],[0,3,5]])
print("")
print(df.loc[[0,90,900],["country","lifeExp","gdpPercap"]])
#해당 방식은 iloc이 아닌 loc만 반환
```
